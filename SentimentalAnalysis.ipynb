```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pyautogui
import pyperclip
import subprocess
from cryptography.fernet import Fernet
import os

pyautogui.hotkey('ctrl', 'v')

# Declaring important coordinates
screenWidth, screenHeight = pyautogui.size()
center_coords = (screenWidth//2, screenHeight//2)

browser_tabs_coords = [(150 + 300*i, 10) for i in range(5)]
search_coords = (720, 950)
copy_button_coords = (764, 871)

# Create a WebDriver instance
driver = webdriver.Chrome()

# URL of the page to scrape
url = "https://www.cnbctv18.com/market/amp"

# Load the page
driver.get(url)

# Find all elements with class name "story-item"
elements = driver.find_elements(By.CLASS_NAME, "story-item")

links = []
for element in elements:
    # Append each link to the list
    links.append(element.find_elements(By.TAG_NAME, 'a')[1].get_attribute("href"))

data = []

# Visit each link and collect text
for link in links[0:3]:
    driver.get(link)
    time.sleep(2)  # Add a short delay to allow the page to load properly

    # Try clicking the "continue" button if present
    try:
        continue_button = driver.find_element(By.ID, "continue-btn")
        continue_button.click()
        time.sleep(2)  # Add a short delay to allow additional content to load
    except:
        pass  # If the button is not found, continue without clicking

    # Find all elements with class name "nart-para"
    text_content = driver.find_elements(By.CLASS_NAME, "nart-para")

    # Collect text from each line and store it as a paragraph
    para = ""
    for line in text_content:
        para += line.text

    data.append(para)

# Store the collected text and add it to the prompt
prompt = ""
for content in data:
    prompt += content + '\n\n'

prompt += "From each of the following paragraphs, find - Name of the stocks, sentiment associated with it as either positive, negative or neutral in one word strictly. Present the output in this format: Stock name - sentiment"
# prompt = prompt.replace('\n', ' ').replace('\r', '')

# Copy the prompt to the clipboard
time.sleep(2)
# Copy the prompt to the clipboard using pyperclip
pyperclip.copy(prompt)

# Close the browser and wait for some time
driver.quit()
time.sleep(1)

# Move the mouse to ChatGPT tab and open it
pyautogui.moveTo(browser_tabs_coords[1], duration=0.6, tween=pyautogui.easeInOutQuad)
pyautogui.click()

# Move the mouse to search bar and click on it to start typing
pyautogui.moveTo(search_coords, duration=0.6, tween=pyautogui.easeInOutQuad)
pyautogui.click()

# Paste the prompt and hit enter
# pyautogui.write(prompt, interval=0.000001)
pyautogui.hotkey('ctrl', 'v')

# Optionally, press Enter to initiate the search
pyautogui.press('enter')

# Wait and then move the mouse to the center of the screen and scroll to the bottom
time.sleep(5)
pyautogui.moveTo(center_coords, duration=0.6, tween=pyautogui.easeInOutQuad)
pyautogui.scroll(-10000)

# # Move to the copy button possible positions and click on them to copy the output
pyautogui.moveTo(copy_button_coords, duration=0.6, tween=pyautogui.easeInOutQuad)
pyautogui.click()
time.sleep(0.5)
res = pyperclip.paste()
print(res)
pyautogui.moveTo(browser_tabs_coords[0], duration=0.6, tween=pyautogui.easeInOutQuad)
pyautogui.click()
# pyautogui.moveTo(copy_button_coords[0], duration=0.6, tween=pyautogui.easeInOutQuad)
# pyautogui.click()

# After copying the data we can paste it into a text file
# Then we can load the dataframe into our code
# And display the names of the stock graphically
# But it is not needed as ChatGPT is also displaying them

# Generate a key for encryption and decryption
key = Fernet.generate_key()
cipher_suite = Fernet(key)

def encrypt_text(text):
    encrypted_text = cipher_suite.encrypt(text.encode())
    return encrypted_text

def decrypt_text(encrypted_text):
    decrypted_text = cipher_suite.decrypt(encrypted_text).decode()
    return decrypted_text

# Encrypt the text
encrypted_text = encrypt_text(res)

# Check if "encode.txt" file exists and delete it if it does
if os.path.exists("encode.txt"):
    os.remove("encode.txt")

# Write the encrypted text to a file in binary mode
with open("encode.txt", "wb") as encode_file:
    encode_file.write(encrypted_text)

# Read the encrypted text from the file in binary mode
with open("encode.txt", "rb") as encode_file:
    encrypted_text_from_file = encode_file.read()

# Decrypt the text
decrypted_text = decrypt_text(encrypted_text_from_file)

# Check if "decode.txt" file exists and delete it if it does
if os.path.exists("decode.txt"):
    os.remove("decode.txt")

# Write the decrypted text to a new file
with open("decode.txt", "w", encoding="utf-8") as decode_file:
    decode_file.write(decrypted_text)
```
